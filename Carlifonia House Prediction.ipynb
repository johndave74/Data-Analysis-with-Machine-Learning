### Feature Engineering

```python
X = np.append(arr = np.ones((1460,1)), values = X, axis=1)
```

```python
import statsmodels.api as sm
X_opt = X[:,[0,1,2,3,4,5]]
regressor = sm.OLS(endog = Y, exog = X_opt).fit()
regressor.summary()
```
